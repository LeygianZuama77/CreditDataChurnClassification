```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# ALAT UNTUK IMPUTASI
from sklearn.impute import SimpleImputer

from sklearn.preprocessing import OneHotEncoder, MinMaxScaler

# Call Model
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier as ada
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier as rforest
from catboost import CatBoostClassifier
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import AdaBoostClassifier as ada
from sklearn.model_selection import GridSearchCV

# Define the numerical preprocessor
num_preprocessor = Pipeline([
    ("num_imputer", SimpleImputer(strategy='mean')),
    ("num_scaler", MinMaxScaler()),
])
cat_preprocessor = Pipeline([("cat_imputer",SimpleImputer(strategy='most_frequent')),
                             ("cat_encoder",OneHotEncoder())])
# Define the numerical columns
num_columns = [
    'usia','jumlah_tanggungan', 'lama_nasabah', 'jumlah_produk', 'bulan_nonactive', 'jumlah_kontak',
    'total_limit_kredit', 'total_limit_kredit_dipakai', 'sisa_limit_kredit', 'rasio_transaksi_Q4_Q1 ',
    'total_transaksi', 'jumlah_transaksi', 'rasio_jumlah_transaksi_Q4_Q1', 'rasio_pemakaian',
    'penghasilan_tahunan_$120K +', 'penghasilan_tahunan_$40K - $60K', 'penghasilan_tahunan_$60K - $80K',
    'penghasilan_tahunan_$80K - $120K', 'penghasilan_tahunan_Less than $40K', 'penghasilan_tahunan_Unknown',
    'tipe_kartu_kredit_Blue', 'tipe_kartu_kredit_Gold', 'tipe_kartu_kredit_Platinum', 'tipe_kartu_kredit_Silver'
]
cat_columns = ['age_category']

# Define the ColumnTransformer with only the numerical pipeline
preprocessor = ColumnTransformer([("num_pipeline",num_preprocessor,num_columns),
                                  ("cat_pipeline",cat_preprocessor,cat_columns)])

# Complete pipeline

adaboost = Pipeline([
    ("preprocess", preprocessor),
    ("model", ada())
])

xgb = Pipeline([
    ("preprocess", preprocessor),
    ("model", XGBClassifier())
])

DecisionTree = Pipeline([
    ("preprocess", preprocessor),
    ("model", DecisionTreeClassifier())
])

catboost = Pipeline([
    ("preprocess", preprocessor),
    ("model", CatBoostClassifier())
])

randomForest =Pipeline([
    ("preprocess", preprocessor),
    ("model", rforest())
])

knn = Pipeline([
    ("preprocess", preprocessor),
    ("model", KNeighborsClassifier())
])

LogistRegress = Pipeline([
    ("preprocess", preprocessor),
    ("model", LogisticRegression())
])
```

```python
# Import Library GSCV
from sklearn.model_selection import GridSearchCV
```

# Imbalance

```python
adaboost_gscv = GridSearchCV(estimator=adaboost,
                          param_grid={
                                'preprocess__num_pipeline__num_imputer__strategy': ['mean', 'median'],
    'preprocess__cat_pipeline__cat_imputer__strategy': ['most_frequent', 'constant'],
    'model__n_estimators': [50, 100, 200],
    'model__learning_rate': [0.01, 0.1, 1.0]
                          },
                          cv=5,
                          verbose=1)

adaboost_gscv.fit(x_train, y_train)
```

```python
adaboost_gscv_balanced = adaboost_gscv.fit(X_train_smotenc,y_train_smotenc)
```

```python
from sklearn.metrics import classification_report

def model_performance(model):
    y_train_pred = model.predict(x_train)
    y_test_pred  = model.predict(x_test)
    print('Model Performance on Data Training')
    print(classification_report(y_train,y_train_pred))

    print('Model Performance on Data Testing')
    print(classification_report(y_test,y_test_pred))
```

```python
def classification_model_performance(model,X,y,label="Training"):
    print("Model Performance in Data",label)
    y_pred = model.predict(X)
    print(classification_report(y,y_pred))
```

Unbalance

```python
model_performance(adaboost_gscv)
```

Balancing

```python
classification_model_performance(adaboost_gscv_balanced,X_train_smotenc,y_train_smotenc,"Training")
classification_model_performance(adaboost_gscv_balanced,x_test,y_test,"Testing")
```
